So far I noticed that many time these characters ('-' and '?') appears in many columns, let quantifies them inside each column of the df:

```python
def count_exact_matches(column):
    return column.isin(['-', '?']).sum()

results = {col: count_exact_matches(df[col]) for col in df.columns}
results_df = pd.DataFrame(list(results.items()), columns=['Column', 'Count'])
results_df
```

Based on this last plot I decide to drop from the data frame both "-" and "?" column since most of the values are inconsistent

```python
df = df.drop(columns=['other fuel type'])
```

Furthermore I need to tranform '-' and '?' characters to null values

```python
df.replace('-', np.NaN, inplace=True)
df.replace('?', np.NaN, inplace=True)
```

Now the rows where the make, the model and the version values are null, will be dropped because cannot be healed

```python
df = df[df['make'].notna()]
df = df[df['model'].notna()]
df = df[df['version'].notna()]
df = df[df['trim'].notna()]
```

Repeat now the check as above of the nul values count per each column:

```python
null_counts = df.isnull().sum()
total_rows = len(df)
null_percentages = (null_counts / total_rows) * 100
null_data = pd.DataFrame({'Column': null_counts.index,
                          'Null_Count': null_counts.values,
                          'Null_Percentage': null_percentages.values})
null_data
```

Now i want to substitute the null values within the three dimensional columns taking the mean by grouping make and model of the cars

```python
# Just to be sure I will plot with the use of three bar charts the distribution of
# the mean values to check if outliers are there:
sns.displot(df, x="length", bins=200, height=2, aspect=5)
sns.displot(df, x="width", bins=200, height=2, aspect=5)
sns.displot(df, x="height", bins=200, height=2, aspect=5)
```

No evident outliers found in the three dimensional columns, so i can proceed with the substitution of the null values by taking the mean values of the group by make and model

```python
# let calculate for each dimensional feature the related mean values by make and model:
length_mean = df.groupby(['make', 'model'])['length'].transform('mean')
width_mean = df.groupby(['make', 'model'])['width'].transform('mean')
height_mean = df.groupby(['make', 'model'])['height'].transform('mean')
```

```python
# Having obtained the mean for each make and model group I will substitute the related missing values by index position:
df['length'] = df['length'].fillna(length_mean)
df['width'] = df['width'].fillna(width_mean)
df['height'] = df['height'].fillna(height_mean)
```

Now I want to heal the missing data present within the column 'engine ltr':

```python
sns.displot(df, x="engine ltr", bins=200, height=2, aspect=5)
```

With this view is possible to see that are not present outliers inside the engine ltr column, despite the presence of some over dimensioned engine

```python
display(df[df['engine ltr'].isnull()])
```

From this subset I can see all the null values remained within the 'engine ltr' column regard a specific car model (Citroen, saxo, elettrica). Now I need to see how dimensioned are the other electric vehicles doing a sub-set:

```python
display(df[df['fuel']=='E'])
```

From this view I can see that Lexus has 10 times more HPs than the CITROEN:SAXO, so I can heal the missing data doing a proportion:

```python
df.loc[(df['model'] == 'CITROEN:SAXO') & (df['engine ltr'].isna()), 'engine ltr'] = 1.1
```

Now I need to heal with the missing data present within the driven wheel column:

```python
display(df[df['model']=='HONDA:FR-V'])
```

Based on the information regarding the driven wheel of the model 'HONDA:FR-V', all the cars with this model present within the data frame present F as values, so I will manage the last null values inserting this value instead.

```python
df.loc[(df['model'] == 'HONDA:FR-V') & (df['driven wheel'].isna()), 'driven wheel'] = 'F'
```

It means that 734 vehicles with a price above 750.000 euro has been dropped from the data frame.
Specifically, most of these vehicles did not have high quality brand such as ferrari, instead they had as brand FIAT, DAEWOO, and so on.

Now I create a new column obtained from the sum of all the volumes of sales occured from january untill december per each year sale (2003, 2004, 2005)

```python
df['Total_Sales'] = df.loc[:, 'MVLJAN':'MVLDEC'].sum(axis=1)
```

Now instead I obtain the number of average sales per month

```python
df['Average_Monthly_Sales'] = df.loc[:, 'MVLJAN':'MVLDEC'].mean(axis=1)
```

Let create new columns starting from existing columns: Power HP, engine ltr, lenght and height

```python
# Let create a new column named sport car for only the cars with high power hp and large engine liters (1, 0):
df['sport_car'] = np.where((df['power (hp)'] >= 300) & (df['engine ltr'] > 3000), 1, 0)

# Let create a second new column named city car for only the cars with low hp power, three doors and small engine liters:
df['city_car'] = np.where((df['power (hp)'] <= 120) & (df['engine ltr'] < 1500) & (df['number doors'] == 3.0), 1, 0)

# Let create a third new column named city car for only the cars with low hp power, three doors and small engine liters:
df['suv_car'] = np.where((df['length'] >= 4500) & (df['height'] > 1600), 1, 0)
```

Let create new columns based on the type of car brands: Such as super cars, high cars, normal cars.

```python
# Create a new column where if the brand is in this list then 1 else 0, for top cars:
top_brands = ['BENTLEY', 'FERRARI', 'MASERATI', 'LAMBORGHINI', 'ASTON MARTIN','MORGAN']
df['top_car'] = df['make'].isin(top_brands).astype(int)

# Create a new column where if the brand is in this list then 1 else 0, for high standard cars:
high_brands = ['MERCEDES','BMW','AUDI','VOLVO','ROVER','JAGUAR','LAND ROVER', 'PORSCHE','JEEP','CADILLAC','LOTUS','LEXUS','HUMMER']
df['high_car'] = df['make'].isin(high_brands).astype(int)
```

Let create four new columns based on the car price targets:

```python
# Car price between 1 and 15,000 Euro for cheap cars:
df['cheap_car'] = np.where((df['price'] > 1) & (df['price'] <= 15000), 1, 0)

# Car price between 15,000 and 30,000 Euro for normal cars:
df['normal_car'] = np.where((df['price'] > 15000) & (df['price'] <= 30000), 1, 0)

# Car price between 30,000 and 100,000 Euro for expensive cars:
df['expensive_car'] = np.where((df['price'] > 30000) & (df['price'] <= 100000), 1, 0)

# Car price over 100,000 for luxury cars:
df['luxury_car'] = np.where(df['price'] > 100000, 1, 0)
```

Reset the final index:

```python
df.reset_index(drop=True, inplace=True)
```
